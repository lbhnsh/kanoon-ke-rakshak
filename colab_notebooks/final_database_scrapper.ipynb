```python
import requests
from bs4 import BeautifulSoup
import textwrap
from fpdf import FPDF
import re
link=''
heading=''
documents = {}
def replace_spaces_and_page_num(query, page_num):
    replaced_query = query.replace(" ", "%20")
    query_with_page_num = f"{replaced_query}&page_num={page_num}"
    return query_with_page_num

def scrape_website(user_query):
    current_page = 0
    headings=[]


    for i in range(5):
        query_with_page_num = replace_spaces_and_page_num(user_query, current_page)
        url = f"https://indiankanoon.org/search/?formInput={query_with_page_num}"  # Replace example.com with the actual website URL
        response = requests.get(url)

        # Check if request was successful
        if response.status_code == 200:
             # Parse the HTML content
            soup = BeautifulSoup(response.content, 'html.parser')

            # Extract result_title and headline divs
            result_titles = soup.find_all('div', class_='result_title')
            headlines = soup.find_all('div', class_='headline')


            # Extract href attributes from <a> tags
            for title, headline in zip(result_titles, headlines):
                heading = title.find('a').text.strip()

                headings.append(heading)
                link = title.find('a')['href']
                documents[heading]=f'https://indiankanoon.org/{link}'
            current_page += 1

    print(headings)
    return documents

# Example usage:
user_query = "forgery"
documents = scrape_website(user_query)
print(documents)

choice=input('Enter choice')

def func(link):
                url1 = requests.get(link1)
                soup1 = BeautifulSoup(url1.text, 'html.parser')

                rb = soup1.find('div',class_='akoma-ntoso')
                if rb==None:
                  rb = soup1.find('div',class_='judgments')
                if rb==None:
                  rb = soup1.find('div',class_='expanded_headline')

                soup1 = BeautifulSoup(str(rb), 'html.parser')
                pdf=FPDF()
                pdf.add_page()
                pdf.set_font('Courier', 'B', 16)

                pdf.cell(0, 10, heading, 0, 1, 'C')
                pdf.set_font('Courier','',16)
                text=soup1.get_text()
                text = text.replace('\u2014', '-')
                text = text.replace('\u201c', '"')
                text = text.replace('\u201d', '"')
                text = text.replace('\u2018', "'")
                text = text.replace('\u2019', "'")

                text.encode('latin-1', errors='ignore')
                print(text)
                pdf.multi_cell(0, 10, text)
                # pdf.output(f'{heading}.pdf','F')
                # Save PDF file
                pdf_output_path = f"{choice}.pdf"
                pdf.output(pdf_output_path, 'F')
                return pdf_output_path
link1=documents[choice]
print(link1)
pdf_path=func(link1)
print(pdf_path)
```
